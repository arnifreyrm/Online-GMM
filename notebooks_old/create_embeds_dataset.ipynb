```python
import numpy as np 
import sys
import os
if "Online-GMM" not in os.getcwd():
    os.chdir("./Online-GMM") # TODO: remove this
data = np.load('data.npy', allow_pickle=True)
```

```python
from dotenv import load_dotenv
import os
load_dotenv()
api_key = os.getenv("OPENAI_API_KEY")
```

```python
import openai
client = openai.Client(api_key=api_key)
```

```python
from functions.utils import get_embedding
```

```python
res = data[0]
embed = get_embedding(client, res)
```

```python
embed[0].embedding
```

```python
from tqdm import tqdm
num_embeds_to_gen = 1000
from openai import BadRequestError

embed_dataset = []
for i in tqdm(range(1, num_embeds_to_gen)):
    text_sample = data[i]
    try:
        embed = get_embedding(client, text_sample)[0].embedding
        embed_dataset.append(embed)
    except BadRequestError as e:
        print(e)
        continue
    
    
```

```python
embed_dataset = np.array(embed_dataset)
# save the embeddings
np.save('embeddings.npy', embed_dataset)
```
